```python
 import cv2
 import numpy as np
 import matplotlib.pyplot as plt
 %matplotlib inline
```

```python

img = cv2.imread('eye.jpg', cv2.IMREAD_UNCHANGED)
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/eye.jpg', cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray_blurred = cv2.blur(gray, (3, 3))
  
```

```python
detected_circles = cv2.HoughCircles(gray_blurred, 
                   cv2.HOUGH_GRADIENT, 1, 20, param1 = 30,
               param2 = 30, minRadius = 1, maxRadius = 20)
```

```python
 if detected_circles is not None:
          detected_circles = np.uint16(np.around(detected_circles))
```

```python
for pt in detected_circles[0, :]:
        a, b, r = pt[0], pt[1], pt[2]
  
```

```python
cv2.circle(img, (a, b), r, (0, 255, 0), 5)
```

```python
cv2.circle(img, (a, b), 1, (0, 0, 255), 3)
```
